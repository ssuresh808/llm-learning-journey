```python
import ollama

# Task 1.1: Simple query
response = ollama.chat(
    model='llama3.2:1b',
    messages=[{'role': 'user', 'content': 'Explain embeddings in one sentence'}]
)
print(response['message']['content'])

# Task 1.2: Temperature experiment
for temp in [0.1, 0.7, 1.5]:
    response = ollama.chat(
        model='llama3.2:1b',
        messages=[{'role': 'user', 'content': 'Write a creative name for a coffee shop'}],
        options={'temperature': temp}
    )
    print(f"Temp {temp}: {response['message']['content']}")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Create embeddings
sentences = [
    "I love programming in Python",
    "Python is my favorite coding language",
    "I enjoy cooking Italian food",
]

embeddings = model.encode(sentences)
print(f"Shape: {embeddings.shape}")

# Calculate similarity
similarity_matrix = cosine_similarity(embeddings)
print("\nSimilarity Matrix:")
print(similarity_matrix)
```
